```python
import matplotlib.pyplot as plt

players=['David warner' , 'KL Rahul', 'Quinton','Shikhar Dhawan','Andre','Kris Gayle','Rishabh Pant','Virat Kohli','Shreyas Iyer', 'Jonny']
scores=[692,593,529,521,510,490,488,464,463,445]
bars=plt.bar(players,scores)
plt.xlabel("Players")
plt.xticks(rotation=45)
plt.ylabel("IPL Scores")
for bar in bars:
    val=bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2,val+5,str(val),ha='center',va='bottom')
plt.legend()
plt.show()
```

```python
import numpy as np
scores=[692,593,529,521,510,490,488,464,463,445]
plt.boxplot(scores)
plt.title("IPL SCORES")
plt.show()
q1=np.percentile(scores,25)
q3=np.percentile(scores,75)
iqr=q3-q1
print("IQR: ",iqr)


```
