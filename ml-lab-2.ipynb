```python
df["Norm"] = df.groupby("ssc_percentage")["hsc_percentage"].transform(
    lambda x: (x - x.mean()) / x.std()
)
```

```python
df["ssc_percentage"] = df["hsc_percentage"].rolling(3).mean()
```

```python
mean = df["ssc_percentage"].mean()
std = df["ssc_percentage"].std()

outliers = df[(df["ssc_percentage"] > mean + 3*std) | (df["ssc_percentage"] < mean - 3*std)]
```
