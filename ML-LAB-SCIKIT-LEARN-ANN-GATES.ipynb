IMPLEMENT AND,OR,NOT AND XOR GATE BY USING ANNS, USE SCIKIT LEARN FOR IT

```python
import numpy as np
from sklearn.neural_network import MLPClassifier
```

```python
# AND GATE
X_and =np.array([[0,0],[0,1],[1,0],[1,1]])
Y_and=np.array([0,0,0,1])

model_and = MLPClassifier(hidden_layer_sizes=(2,), max_iter=1000)
model_and.fit(X_and, Y_and)

print("AND Gate:")
print(model_and.predict(X_and))
```

```python
X_or = np.array([[0,0],[0,1],[1,0],[1,1]])
y_or = np.array([0,1,1,1])
model_or = MLPClassifier(hidden_layer_sizes=(2,), max_iter=5000)
model_or.fit(X_or, y_or)

print("OR Gate:")
print(model_or.predict(X_or))
```

```python
X_xor = np.array([[0,0],[0,1],[1,0],[1,1]])
y_xor = np.array([0,1,1,0])
model_xor = MLPClassifier(hidden_layer_sizes=(4,), max_iter=2000)
model_xor.fit(X_xor, y_xor)

print("XOR Gate:")
print(model_xor.predict(X_xor))
```

```python
X_not = np.array([[0],[1]])
y_not = np.array([1,0])
model_not = MLPClassifier(hidden_layer_sizes=(2,), max_iter=1000)
model_not.fit(X_not, y_not)
print("NOT Gate:")
print(model_not.predict(X_not))
```
