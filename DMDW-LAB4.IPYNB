```python
#TASK 1
import numpy as np
import scipy.stats
from scipy import stats
a=np.array([30,36,47,50,52,52,56,60,63,70,70,110])
print(np.mean(a))
print(np.median(a))
print(stats.mode(a))
```

```python
#TASK 2
b=np.array([5, 8, 3, 4, 7, 12, 5, 2])
c=max(b)
d=min(b)
print("range: ",c-d)
print(np.var(b))
print(np.std(b))
q1=np.percentile(b,25)
q3=np.percentile(b,75)
print("IQR= ",q3-q1)
```
