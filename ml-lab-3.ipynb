Scatter and Bar Plots

```python
x = [1,2,3,4,5]
y = [5,7,8,7,6]

plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,color='red',s=100)
plt.show()
```

```python
categories = ['A','B','C','D']
values = [10,20,15,25]

plt.bar(categories,values)
plt.show()
```

```python
plt.barh(categories,values)
plt.show()
```

```python
bars = plt.bar(categories,values)

for bar in bars:
    plt.text(bar.get_x()+0.2, bar.get_height()+0.5,
             str(bar.get_height()))

plt.show()
```

```python
x = np.arange(3)

boys = [20,25,30]
girls = [22,28,26]

plt.bar(x-0.2,boys,width=0.4,label='Boys')
plt.bar(x+0.2,girls,width=0.4,label='Girls')

plt.legend()
plt.show()
```

```python
categories = ['A','B','C']   
boys = [20,25,30]            
girls = [22,28,26]          
plt.bar(categories, boys, label="Boys")
plt.bar(categories, girls, bottom=boys, label="Girls")
plt.legend()
plt.show()
```

```python
categories = ['Apple','Banana','Orange','Mango']
values = [10,15,7,12]
plt.bar(categories, values)      
plt.xticks(rotation=45)          
plt.show()             
```

```python
plt.bar(categories,values,width=0.3)
plt.show()
```

```python
plt.bar(categories,values,color=['red','green','blue','orange'])
plt.show()
```

Histogram and Box Plot

```python
data = np.random.randn(1000)
plt.hist(data)
plt.show()
```

```python
plt.hist(data,bins=20)
plt.show()
```

```python
plt.hist(data,bins=20,density=True)
plt.show()
```

```python
data2 = np.random.randn(1000)
plt.hist(data,alpha=0.5,label="Data1")
plt.hist(data2,alpha=0.5,label="Data2")
plt.legend()
plt.show()
```

```python
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data, patch_artist=True,
            boxprops=dict(facecolor='lightblue'))
plt.show()
```

```python
plt.boxplot([data,data2])
plt.show()
```

```python
plt.boxplot(data,showfliers=True)
plt.show()
```

```python
plt.boxplot([data,data2],labels=['Data1','Data2'])
plt.show()
```

```python
plt.tight_layout()
plt.show()
```

Subplots and Layouts

```python
fig, ax = plt.subplots(2,2)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
categories = ['A','B','C','D']
values = [10,15,7,12]
data = np.random.randn(100)
fig, ax = plt.subplots(2,2)
ax[0,0].plot(x,y)
ax[0,1].scatter(x,y)
ax[1,0].bar(categories,values)
ax[1,1].hist(data)
plt.show()
```

```python
fig, ax = plt.subplots(2,2,sharex=True,sharey=True)
plt.subplots_adjust(wspace=0.5,hspace=0.5)
ax[0,0].set_title("Line Plot")
plt.suptitle("Multiple Plots")
plt.savefig("subplot.png")
```

```python
fig.delaxes(ax[1,1])
```

```python
plt.tight_layout()
```

```python
ax[0,0].set_facecolor('lightgray')
```

```python
plt.savefig("subplot.png")
```

```python
dates = pd.date_range('2024-01-01', periods=10)
values = np.random.randint(10,50,10)

plt.plot(dates,values)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.plot(x,y)
plt.annotate("Peak",(5,10))
plt.show()
```

```python
data = pd.Series(np.random.randn(50))
ma = data.rolling(window=5).mean()

plt.plot(data,label='Original')
plt.plot(ma,label='Moving Avg')

plt.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x,y,color='blue')

ax2 = ax1.twinx()
ax2.plot(x,y2,color='red')

plt.show()
```

```python
x = np.random.rand(100)
y = x*2 + np.random.rand(100)

plt.scatter(x,y)
plt.show()
```

```python
mean = np.sin(x)
std = 0.2

plt.plot(x,mean)
plt.fill_between(x, mean-std, mean+std, alpha=0.3)

plt.show()
```

```python
plt.plot(x,y)
plt.xticks([1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])
plt.show()
```

```python
plt.fill_between(x,y,alpha=0.4)
plt.show()
```
