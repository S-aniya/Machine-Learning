```python
import matplotlib.pyplot as plt
```

```python
x=[15,50,90,115,140,190]
a=[2017, 2018, 2019, 2020, 2021, 2022]
y=[80,50,100,75,95,70]
z=[150,75,30,10,5,0]
plt.plot(a,x,color='blue',marker='o')
plt.plot(a,y,color='grey',marker='o')
plt.plot(a,z,color='orange',marker='o')
plt.legend()
plt.show()
```

```python
colors = ['Red', 'Blue', 'Green', 'Yellow', 'Pink']
children = [8, 7, 5, 9, 6]

bar_colors = ['red','blue', 'green', 'yellow', 'pink']

plt.bar(colors, children, color=bar_colors)

plt.ylabel("Children")

plt.show()
```

```python
bins=[20,30,40,50,60,70,80.90,100]
frequency=[2,.4,4,5,3,1,0,1]
plt.bar(bins,frequency,width=10)
plt.show()
```
