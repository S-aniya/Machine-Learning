Part B: Scatter & Bar Plots

```python
#Create a scatter plot
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,color='pink',s=100,marker='s')
plt.show()
```

```python
plt.bar(x,y,color='pink')
plt.show()
```

```python
plt.barh(x,y2,color='pink')
plt.show()
```

```python
plt.bar(x,y)
for i in range(len(x)):
    plt.text(x[i],y[i],y[i])
plt.show()
```

```python
import numpy as np
x=np.arange(3)
a=[3,5,7]
b=[4,6,8]
plt.bar(x-0.2,a,width=0.4,label="Group A")
plt.bar(x+0.2,b,width=0.4,label="Group B")
plt.legend()
plt.show()
```

```python
plt.bar(x, a, label="A")
plt.bar(x, b, bottom=a, label="B")
plt.legend()
plt.show()
```

```python
plt.bar(x, a)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(x, a, width=0.3)
plt.show()
```

```python
plt.bar(x, a, color='blue')
plt.show()
```

PART C: HISTOGRAM & BOX PLOT

```python
data=np.random.randn(100)
plt.hist(data)
plt.show()
```

```python
plt.hist(data, bins=20)
plt.show()
```

```python
plt.hist(data, density=True)
plt.show()
```

```python
plt.hist(data)
plt.hist(data + 1)
plt.show()
```

```python
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data, patch_artist=True)
plt.show()
```

```python
plt.boxplot([data, data + 2])
plt.show()
```
